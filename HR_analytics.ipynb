```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing Data

```python
data = pd.read_csv('HR_ANALYTICS.csv')
```

```python
pd.options.display.max_columns
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
data.columns = data.columns.str.lower().str.replace('-', '_')
data.head()
data.columns = data.columns.str.replace(' ', '_')
data.head()
```

```python
data.isna().sum()
```

# Cleaning & Dealing with NA's

```python
mnn = data['duration_to_accept_offer'].mean()
round(mnn)
```

```python
data['duration_to_accept_offer'] = data['duration_to_accept_offer'].fillna(mnn)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

# Exploratory Data Analysis EDA

```python
data.groupby('d_joining').mean()
```

```python
corr = data.corr()
```

```python
corr['d_joining'].sort_values(ascending=True)
```

```python
x = data[['d_doj_extended_','duration_to_accept_offer','d_infra',
          'percent_hike_offered_in_ctc','percent_difference_ctc','age']]
x
```

```python
y = data['d_joining']
y
```

# Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
y_predlr = lr_model.predict(X_test)
```

```python
y_predlr
```

```python
from sklearn.metrics import accuracy_score,mean_squared_error
```

```python
ac = accuracy_score(y_predlr,y_test)*100
```

```python
ac
```

```python
er = mean_squared_error(y_predlr,y_test)*100
```

```python
er
```

```python
lr_model.predict([[0,5,0,10,-3,50]])
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_predlr))
```

```python
print(confusion_matrix(y_test,y_predlr))
```

# Decision Tree Model

```python
from sklearn import tree
```

```python
dtmodel = tree.DecisionTreeClassifier(random_state = 42)
```

```python
dtmodel.fit(X_train,y_train)
```

```python
dtmodel.predict([[10,3,0,4,-2,26]])
```

```python
y_dt_pred = dtmodel.predict(X_test)
```

```python
y_dt_pred
```

```python
accuracy_score(y_dt_pred,y_test)*100
```

```python
mean_squared_error(y_dt_pred,y_test)*100
```

```python
print(classification_report(y_dt_pred,y_test))
```

```python
print(confusion_matrix(y_dt_pred,y_test))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model = RandomForestClassifier(n_estimators = 20,
                                  random_state= 42)
```

```python
rfc_model.fit(X_train,y_train)
```

```python
rfc_model.predict([[10,3,1,45,-45,56]])
```

```python
y_pred_rfc = rfc_model.predict(X_test)
```

```python
y_pred_rfc
```

```python
accuracy_score(y_pred_rfc,y_test)*100
```

```python
mean_squared_error(y_pred_rfc,y_test)*100
```

```python
print(classification_report(y_pred_rfc,y_test))
```

```python
print(confusion_matrix(y_pred_rfc,y_test))
```

```python
rfc_model.estimators_
```
